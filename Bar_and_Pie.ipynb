```python
import pandas as pd

# Load the dataset
file_path = 'C:/Users/Hari/Desktop/DataVisualization/Netflix Userbase.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Bar Plot: Count of users per country
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
data['Country'].value_counts().plot(kind='bar')
plt.title('Number of Users per Country')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.show()
```

```python
# Pie Chart: Proportion of users by subscription type
plt.figure(figsize=(5,5))
data['Subscription Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Users by Subscription Type')
plt.ylabel('')
plt.show()
```
